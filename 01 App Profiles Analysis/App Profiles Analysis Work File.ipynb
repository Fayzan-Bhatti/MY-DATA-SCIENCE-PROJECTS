```python
def clean_eng(app):
    #print(app)
    non_eng=0
    for i in app:
        #print(i)
        if ord(i)>128:
            non_eng += 1
    if non_eng>3:
        return False
    else:
        return True
        
```

```python
clean_eng("Faceboodajdidjaisdjasioadj<(a#k >()😍")
```

```python
clean_eng("Faceboodajdidjaisdjasioadj<(a#k >()😍😍😍😍😍😍")
```
