# Merging 

* merge means merg the data on someonce commen point that you define 
* merge always work on horizontal 
* while concate works in both horizontal and vertical position
* we merge the data at commen point, Ans DBN is commen point
* in pandas merge=join
* in merging we join just 2 data once time
* in database we have 2 join >>>left and inner
* but basically we have 4 join
* left , right , inner , outer
* left join mean data on left side complete and on right side just merge tha data that key match
* right join mean data on right side complete and on left side just merge tha data that key match
* inner join = commen data in both right and left side= intersect
* outer join = all data (once time) in both right and left side=union
*
