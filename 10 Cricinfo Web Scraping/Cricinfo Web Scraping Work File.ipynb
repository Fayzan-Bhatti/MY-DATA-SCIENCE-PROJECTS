```python
odis=all_tables[1]
odis
```

```python
update_date=odis.caption.text
update_date
```

```python
odis.find_all("th")
```

```python
col=[]
for i in odis.find_all("th"):
    #print(i.text)
    col.append(i.text)
```

```python
col
```

```python
table=odis.find_all("td")
table
```

```python
table_lst=[]
start=0
end=5
for i in range(20):
    pos,team,matches,points,rating=table[start:end]
    table_lst.append([pos.text,team.text,matches.text,points.text,rating.text])
    start += 5
    end += 5
```

```python
table_lst
```

```python
odi_ranking=pd.DataFrame(data=table_lst,columns=col)
odi_ranking
```

```python
odi_ranking.to_csv("odi_ranking.csv",index=False)
```
