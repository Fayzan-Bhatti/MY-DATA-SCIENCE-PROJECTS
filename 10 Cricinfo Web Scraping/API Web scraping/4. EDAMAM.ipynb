```python
import json
import requests
import pandas as pd
```

```python
APP_ID = "f1869549"
APP_KEYS = "acac3d25dea4bf3ca2ed363a4e4925dc"
```

```python
api_endpoint = "https://api.edamam.com/api/nutrition-details"
```

```python
url = api_endpoint + "?app_id=" + APP_ID + "&app_key=" + APP_KEYS
url
```

Sending a POST request

```python
headers = {"Content-Type": "application/json"}
```

```python
receipe = {
    "title" : "Cappuccino",
    "ingr" : ["18g ground espresso", "150ml milk"]
}
```

```python
r = requests.post(url, headers = headers, json = receipe)
r.status_code
```

```python
capp_info = r.json()
print(json.dumps(capp_info, indent = 4))
```
